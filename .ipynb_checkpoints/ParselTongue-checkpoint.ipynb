# Vectorizing Text

The purpose of tokenizing the text is to convert it into a structured format suitable for analysis. This effectively transform a document into a bag of words. The assumption is that the sentence structure of the document (e.g. order of the words) does not play a role in the analysis.

```python
# Showing the result of vectorizing the first 3 reviews in the dataset.
num_reviews = 3
vectorized = pd.DataFrame()
for doc in tokens[:num_reviews]:
    word_dict = {}
    for t in doc:
        word_dict[t] = word_dict.get(t, 0) + 1
    temp = pd.DataFrame([word_dict])
    vectorized = vectorized.append(temp)
    
vectorized = vectorized.fillna(0)
vectorized.index = data.text.iloc[:num_reviews]
vectorized.iloc[:, np.random.randint(0, len(vectorized.columns), 10)]
```

In the vectorized version of the reviews, there will contain punctuations, and words like 'a', 'and', 'will' called stop words. In general, stop words do not convey much meaning to the sentence. Consider removing these words.

# Sparse Matrix

After removing stopwords and lemmatization, the number of words has reduced. However, you will notice many zeros in the matrix. This is because there will be many words contained in some documents, but not in others. 

Note, it is inefficient to save information about no information, i.e. the zeros. A sparse matrix is used instead. For each document, the information stored is a list of **(word column index, word count)**.

E.g. If the word **'look' is in the first column of the matrix (index == 0)**. A document contains **3 counts** of the word 'look'. This word in that document will be saved as **(0, 3)**

# Similarity between documents

Calculating similarity between documents can help us to group documents. Why do we want to do that?

Applications:
1. Plagarism checker:
    * An essay submitted can be compared for similarity with the other essays in the database.
* An online retailer searching for its products that belong in the same category and groups them together. 
    * Helps visitor to its website find similar items.

## Cosine similarity between 2 documents

Document v = (1, 1, 1, $\cdots$, 0) --> $\begin{pmatrix}v_1\\v_2\\v_3\\\vdots\\v_n\end{pmatrix}$ --> $\begin{pmatrix}1\\1\\1\\\vdots\\0\end{pmatrix}$


Document w = (0, 0, 1, $\cdots$, 0) --> $\begin{pmatrix}w_1\\w_2\\w_3\\\vdots\\w_n\end{pmatrix}$ --> $\begin{pmatrix}0\\0\\1\\\vdots\\0\end{pmatrix}$

## $\cos similarity = \frac{v.w}{\|v\|\|w\|}$

### $v.w = v_1 * w_1 + v_2 * w_2 + ... + v_N * w_N$

### $\|v\| = \sqrt{v_1^2 + v_2^2 + ... + v_N^2}$

### $\|w\| = \sqrt{w_1^2 + w_2^2 + ... + w_N^2}$

Invert each document in the matrix, and you'll get the vector of a document. 

v and w are the vectors of 2 different documents.

## A contrived example

Say we want to find the similarity between 3 documents.

1. Restaurant Restaurant Restaurant Restaurant Restaurant Restaurant Restaurant avocado chicken delicious.
* Food delicious.
* Restaurant Restaurant Restaurant Restaurant Restaurant Restaurant Restaurant ambient tempature too cold.

```python
doc1 = "Restaurant Restaurant Restaurant Restaurant Restaurant Restaurant Restaurant chicken pie deliciou"
doc2 = "Restaurant Food deliciou"
doc3 = "Restaurant Restaurant Restaurant Restaurant Restaurant Restaurant Restaurant ambient tempature too cold"

tokens = [d.split() for d in [doc1, doc2, doc3]]
contrive_dict = corpora.Dictionary(tokens)
sparse_vector = [contrive_dict.doc2bow(doc) for doc in tokens]
index = similarities.SparseMatrixSimilarity(sparse_vector, len(contrive_dict.items()))

for i, similarity in enumerate(index[sparse_vector[0]][1:], 1):
    print 'doc1 similarity with doc%i: %f' %(i+1, similarity)
```

Actually, it would seem like doc1 is more similar to doc2 than to doc3

## Adjusting word frequency before performing similarity

**idf** is a measure to multiply to the frequency of words (tf). The result result value is called **tf-idf** (term frequency - inverse document frequency). It's main purpose serves to reduce the frequency of words common in many documents in the corpus, and to increase the frequency of words uncommon in the corpus. Similar to stopwords, words occuring in many documents does not provide good discriminating value between documents, hence its frequency is weigh down.

The formula for idf for each word is:

### $\log_2\frac{D}{df_w}$

where $D$ is the number of documents in the corpus, and $df_w$ is the number of documents containing that word.

Example: There are 10000 documents in the corpus. The word 'HPC' is in 8151 documents, the idf of 'HPC' is:

```python
df_w = 8151
D = 10000

idf_HPC = np.log(D/df_w)/np.log(2)

print idf_HPC
```

Another word, 'bicycle' is only appears in 100 documents.

```python
df_w = 100

idf_bicycle = np.log(D/df_w)/np.log(2)

print idf_bicycle
```

Back to our contrive example:

```python
tfidf_contrieved = models.TfidfModel(sparse_vector)
tfidf_sparse = [tfidf_contrieved[doc] for doc in sparse_vector]
tfidf_index = similarities.SparseMatrixSimilarity(tfidf_sparse, len(contrive_dict.items()))

for i, similarity in enumerate(tfidf_index[tfidf_sparse[0]][1:], 1):
    print 'doc1 similarity with doc%i: %f' %(i+1, similarity)
```

doc1's similarity with doc3 becomes 0, because "delicious" has become a "stopword" in the corpus.

# Topic Modeling

The purpose is to discover the topics in a corpus, and to differentiate between the documents.

As in the vast majority of all text analytics technique, the concept of topics is based on probabilistic models and co-occurance of words. These algorithms do not take into account the semantics (meaning) of words. 

One commonly used topic modeling algorithm is called the Latent Dirichlet Allocation (LDA). We initialize the topic model by assuming there are $X$ number of topics in the corpus, where each topic is characterized by the words it contains. 

Here is an example of LDA on the reviews on the restaurant business. Again, we will be using adjusting the weight of the word by tf-idf.

```python
cat = 'Public Services & Government'
tokens = np.array(reviews)[np.array(data.categories==cat)]
dictionary = corpora.Dictionary(tokens)
sparse_vector = [dictionary.doc2bow(doc) for doc in tokens]
tfidf = models.TfidfModel(sparse_vector)
sparse_w_tfidf = [tfidf[review] for review in sparse_vector]

num_reviews = len(sparse_vector)
print 'There are %i reviews in the "%s" category.' %(num_reviews, cat)
```

```python
PSG_lda = gensim.models.ldamodel.LdaModel(corpus = sparse_w_tfidf, id2word = dictionary, num_topics=20)
PSG_lda.show_topics(5, 8)
```

```python
j = np.random.randint(len(sparse_w_tfidf))

print data.text[data.categories==cat].iloc[j], '\n'

related_topics = PSG_lda[sparse_w_tfidf[j]]
related_topics_sorted = sorted(related_topics, key=lambda x: x[1], reverse=True)

most_relevant = related_topics_sorted[0][0]

print 'The most relevant topic has a probability of', related_topics_sorted[0][1]
print PSG_lda.show_topic(most_relevant, topn=20)
```
