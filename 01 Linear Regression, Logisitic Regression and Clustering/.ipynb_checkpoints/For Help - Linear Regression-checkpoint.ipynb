## Hypothesis

$h(x^{(i)}) = \theta_0 x_0^{(i)} + \theta_1 x_1^{(i)} + \theta_2 x_2^{(i)} + ... + \theta_n x_n^{(i)}$

Expressing the above in matrix notation ...

$\theta =
\begin{bmatrix} \theta_0 \\\theta_1 \\\theta_2 \\\vdots \\\theta_n \\\end{bmatrix}$
$$$$
$x = \begin{bmatrix} x_0^{(1)}&x_1^{(1)}& x_2^{(1)}& \cdots &x_n^{(1)} \\x_0^{(2)}&x_1^{(2)}& x_2^{(2)}& \cdots &x_n^{(2)}\\
\vdots&\vdots&\vdots&\ddots&\vdots\\x_0^{(m)}&x_1^{(m)}& x_2^{(m)}& \cdots &x_n^{(m)}\end{bmatrix}$

To calculate the hypothesis of all the data points, this is simply a matrix multiplication of $\mathbf x$ with $\mathbf \theta$.

## Initialize $\theta_j$

```python
# theta_j, aka parameters
initial_parameters = np.zeros(5)
```

## Cost Function
<br>$J = \frac{1}{2m} \sum_{i=0}^m (y - h(x^{(i)}))^2$<br><br> $where\; y - h(x^{(i)}) = \;error\; of\;prediction\;for\;a\;single\;data\;point$ $$$$

```python
# Make predictions with the parameters and calculate the errors in the predictions.

# some matrix multiplication here
h_x = x.dot(initial_parameters)
print 'With the parameters all initialized to 0, the predictions will be zero as well.'
print h_x[:5]
```

```python
# Defining a function to calculate the cost of the prediction.
def cost(y, h_x):
    errors = y - h_x
    m = len(y)
    
    return sum(errors**2) / (2.0 * m)
```

## Calculate the gradient of each parameter, $\theta_j$

$\frac{\partial J}{\partial\theta_j} = \frac{1}{m} \sum_{i=0}^{m}(y - h(x^{(i)}))(-x_j^{(i)}) $ $$$$

```python
def gradients(x, y, h_x):
    x = np.array(x)
    m = len(y)
    n = x.shape[1] # the number of variables in x (inclusive of the bias term)
    error = y - h_x
    
    slopes = np.zeros(n)
        
    # calculate the gradient of each parameter
    for j in xrange(n):
#         slopes[j] = error.dot(-x[:, j]) / m
#         print 'error', np.array(error).shape
#         print 'x[:, j]', x[:, j].shape
        slopes[j] = np.array(error).dot(-x[:, j]) / m
    
    return slopes
```

## Update $\theta_j$ by a small amount 
$\theta_j = \theta_j - \alpha\frac{\partial J}{\partial\theta_j}$

```python
def parameter_update(parameters, gradients, alpha = 0.1):
    return parameters - float(alpha) * gradients
```

## Step 6 Repeat step 3 to 5

```python
def linear_regression(x, y, initial_parameters, alpha = 1., iter = 100, show='Y'):
    parameters = initial_parameters
    costs = [] # this is for recording the cost a each iteration
    # for logging how the parameters change
#     parameters_list = []
    
    # Step 6
    for i in xrange(iter):
        if i % 500 == 0 and show == 'Y':
            print 'Iteration', i
        # Step 3
        h_x = x.dot(parameters)
        costs.append(cost(y, h_x))
        
        # Step 4
        slopes = gradients(x, y, h_x)
        
        # Step 5
        parameters = parameter_update(parameters, slopes, alpha)
        
        # for logging how the parameters change
#         parameters_list.append(parameters)
    if show == 'Y':
        print 'Completed all', iter, 'iterations.'
    
    
    return parameters, costs#, parameters_list
```

## An alpha that works

```python
parameters, costs = (linear_regression(x, y, initial_parameters, alpha = 6e-7, iter = 50000, show = None))
# Here, I have chosen an alpha that works
```

So the equation for making predictions is ...

```python
print 'PE =',
for j, p in enumerate(parameters):
    if j < (len(parameters) - 1):
        print str(p), '*', features[j], '+',
    else:
        print str(p), '*', features[j]
```

## Check Whether the Algorithm is Implemented Correctly.

If implemented correctly, the cost should decrease with every iteration.

```python
def plot_cost(x, y, title='', ylab='Cost', label = None, legend=None, color = None):
    plt.plot(x, y, color=color, label = label)
    plt.ylabel(ylab)
    plt.xlabel('Iteration')
    plt.title(title)
    if legend != None:
        plt.legend(loc=legend)
    
plot_cost(range(10, len(costs)), costs[10:], title = 'Plot of Cost against Iteration', ylab='Cost')
```

## See the prediction on new data

```python
new_data = pd.read_excel(filename, sheetname=1)

new_data['x_0'] = 1
x_new = new_data[features][:1000]
y_new = new_data[outcome][:1000]
```

```python
predictions = x_new.dot(parameters)

plt.scatter(x_new['AT'], y_new, alpha = 1, color='blue', label='actual')
plt.scatter(x_new['AT'], predictions, color='red', alpha = 0.5, label='predicted')
plt.legend(loc='upper right')
plt.title('Plot of PE against AT')
plt.ylabel('PE')
plt.xlabel('AT')
plt.show()
```

## Larger learning rate

Note: 10 times lesser iteration

```python
print 'When the learning rate is too large.'
alpha = 6e-6 # 10 times less iteration then before. 6e-7 was the alpha used just now
parameters_, costs_ = linear_regression(x, y, initial_parameters, alpha, iter = 5000, show = None) 
plot_cost(range(len(costs_)), costs_, title = 'Plot of Cost against Iteration at alpha = %f' % alpha)
```

A learning rate that is too large results in the cost function increasing without control

## Smaller learning rate

```python
alphas = [6e-7, 3e-7, 6e-8, 3e-8, 6e-9, 3e-9]
iterations = 5000
min_iter = 0

for alpha in alphas:
    initial_parameters = np.zeros(x.shape[1])
    parameters, costs = (linear_regression(x, y, initial_parameters, alpha = alpha
                                                            , iter=iterations, show = None))
    (plot_cost(range(min_iter, iterations), np.log(costs[min_iter:])
               , title='Plot of Cost Against Iteration\n(for different alphas)'
               , ylab='log(Cost)', label = 'alpha = %s' %alpha))

plt.legend(loc='upper right')
plt.show()
```

If the learning rate is too small, the cost will take a long time to stabilize.
