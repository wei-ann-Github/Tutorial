Linear regression can be broken down into these steps.

**Step 1** Start with hypothesis<br><br> $h(x^{(i)}) = \theta_0 x_0^{(i)} + \theta_1 x_1^{(i)} + \theta_2 x_2^{(i)} + ... + \theta_n x_n^{(i)}$$$$$

**Step 2** Give $\theta_j$ an initial value, for example, $\theta_0 = 0$, $\theta_1 = 0$, $\theta_2 = 0$ etc...$$$$
**Step 3** Calculate the cost of the prediction, <br><br>$J = \frac{1}{2m} \sum_{i=0}^m (h(x^{(i)}) - y)^2$<br><br> $where\; h(x^{(i)}) - y = \;error\; of\;prediction\;for\;a\;single\;data\;point$ $$$$

```python
(Image
 .open('C:\Users\limweia\OneDrive - Hewlett Packard Enterprise 1\Projects\Giving Lessons\Linear Regression\img\cost function.png')
 .resize((480, 400)) )
```

**Step 4** Calculate the gradient of each parameter, $\theta_j$: <br><br>$\frac{\partial J}{\partial\theta_j} = \frac{1}{m} \sum_{i=0}^{m}(h(x^{(i)})-y)(x_j^{(i)})\qquad \forall j$ $$$$

```python
(Image
 .open('C:\Users\limweia\OneDrive - Hewlett Packard Enterprise 1\Projects\Giving Lessons\Linear Regression\img\gradient.png')
 .resize((480, 400)) )
```

**Step 5** Update $\theta_j$ by a small amount $$\theta_j = \theta_j - \alpha\frac{\partial J}{\partial\theta_j}$$ $$where\; \alpha\; is\; the\; learning\; rate$$$$$$
**Step 6** Repeat step 3 to 5 until the change in the cost function, $J$, is small enough, or after a certain number of iterations

```python
# Prepare the data
data['x_0'] = 1 # introducing the bias term
features = ['x_0'] + feats

x = data[features]
y = data[outcome]

print 'These are the features:'
print x.head(10)
print '\nThese are the outcomes:'
print y.head(10)

x = np.array(x)
y = np.array(y)
```

## Step 1 Hypothesis

$h(x^{(i)}) = \theta_0 x_0^{(i)} + \theta_1 x_1^{(i)} + \theta_2 x_2^{(i)} + ... + \theta_n x_n^{(i)}$

Expressing the above in matrix notation ...

$\theta =
\begin{bmatrix} \theta_0 \\\theta_1 \\\theta_2 \\\vdots \\\theta_n \\\end{bmatrix}$
$$$$
$x = \begin{bmatrix} x_0^{(1)}&x_1^{(1)}& x_2^{(1)}& \cdots &x_n^{(1)} \\x_0^{(2)}&x_1^{(2)}& x_2^{(2)}& \cdots &x_n^{(2)}\\
\vdots&\vdots&\vdots&\ddots&\vdots\\x_0^{(m)}&x_1^{(m)}& x_2^{(m)}& \cdots &x_n^{(m)}\end{bmatrix}$

To calculate the hypothesis of all the data points, this is simply a matrix multiplication of $\mathbf x$ with $\mathbf \theta$.

## Step 2 Initialize $\theta_j$

```python
# theta_j, aka parameters
initial_parameters = np.ones(5)

print 'Initial Parameters', initial_parameters
```

## Step 3 Calculate the cost of the prediction
<br>$J = \frac{1}{2m} \sum_{i=0}^m (h(x^{(i)}) - y)^2$<br><br> $where\; h(x^{(i)}) - y = \;error\; of\;predictiing\;a\;single\;data\;point$ $$$$

```python
# Make predictions with the parameters and calculate the errors in the predictions.
# Matrix multiplication
h_x = x.dot(initial_parameters)

print h_x[:5]
```

```python
# Defining a function to calculate the cost of the prediction. We'll be using this later.
def cost_function(y, h_x):
    errors = h_x - y
    m = len(y)
    
    return errors.dot(errors) / (2.0 * m)

print 'The predictions and errors of the first 10 rows of data.'
print 'pred\terrors'
for i, pred in enumerate(h_x[:10]):
    print '%.1f' %pred, '\t', pred - y[i]
```

## Step 4 Calculate the gradient of each parameter, $\theta_j$

$\frac{\partial J}{\partial\theta_j} = \frac{1}{m} \sum_{i=0}^{m}(h(x^{(i)}) - y)(x_j^{(i)}) $ $$$$

```python
def gradients(x, y, h_x):
    m = x.shape[0]
    n = x.shape[1] # the number of variables in x (inclusive of the bias term)
    errors = h_x - y
    
    slopes = np.zeros(n)
    
    # calculate the gradient of each parameter
    for j in xrange(n):
        slopes[j] = errors.dot(x[:, j]) / m
    
    return slopes
```

## Step 5 Update $\theta_j$ by a small amount 
$\theta_j = \theta_j - \alpha\frac{\partial J}{\partial\theta_j}$

Choosing that small amount, represented by $\alpha$, is tricky. Anyway, let's define a function for the $\theta$ update and assign 1 as the default value of alpha.

```python
def parameter_update(parameters, gradients, alpha = 0.1):
    return parameters - float(alpha) * gradients
```

## Step 6 Repeat step 3 to 5

```python
def linear_regression(x, y, initial_parameters, alpha = 1., iter = 100, show='Y'):
    parameters = initial_parameters
    costs = [] # this is for recording the cost a each iteration
    
    # Step 6
    for i in xrange(iter):
        if i % 500 == 0 and show == 'Y':
            print 'Iteration', i
        # Step 3
        h_x = x.dot(parameters)
        costs.append(cost_function(y, h_x))
        
        # Step 4
        slopes = gradients(x, y, h_x)
        
        # Step 5
        parameters = parameter_update(parameters, slopes, alpha)
        
    if show == 'Y':
        print 'Completed all', iter, 'iterations.'
    
    return parameters, costs#, parameters_list
```

## Let's start learning

```python
alpha = 1e-6

parameters, costs = linear_regression(x, y, initial_parameters, alpha = alpha, iter = 50000) 
# Here, I have chosen an alpha that works
```

So the equation for making predictions is ...

```python
print 'PE =',
for j in xrange(len(parameters)):
    if j < (len(parameters) - 1):
        print str(parameters[j]), '*', features[j], '+',
    else:
        print str(parameters[j]), '*', features[j]
```

## Check Whether the Algorithm is Implemented Correctly.

If the algorithm is implemented correctly, the cost should decrease with every iteration.

```python
def plot_cost(x, y, title='', ylab='Cost', label = None, legend=None, color = None):
    plt.plot(x, y, color=color, label = label)
    plt.ylabel(ylab)
    plt.xlabel('Iterations')
    plt.title(title)
    if legend != None:
        plt.legend(loc=legend)
    
plot_cost(range(1, len(costs)), costs[1:], title = 'Plot of Cost against Iteration', ylab='Cost')
```

## Prediction in Action

```python
num_data = 1000
predictions_train = x[:num_data,:].dot(parameters)

plt.scatter(x[:num_data,1], y[:num_data], alpha = 1, color='blue', label='actual')
plt.scatter(x[:num_data,1], predictions_train, color='red', alpha = 0.5, label='predicted')
plt.title('Plot of PE against AT (training data)')
plt.legend(loc='upper right')
plt.ylabel('PE')
plt.xlabel('AT')
plt.show()

print 'The cost is:', cost_function(y[:num_data], predictions_train)
```

**IMPORTANT!**

We typically build different models, each giving different predictions on the same $x$. We can compare their performance using the cost_function since the cost_function takes an average of errors of all the data.

However, to gain an estimate on the real world performance of the model developed, we cannot compare the cost on the same data we used to fit the model. This is because our model is optimized on these data. We can build a model that can predict all the y values perfectly. This is called **overfitting** on the data, because we know the real world does not work like this. 

We have to use a fresh set of data for an estimate of the real world performance of this model. The next 2 cells does that.

```python
new_data = pd.read_excel(filename, sheetname=1)

new_data['x_0'] = 1
x_new = np.array(new_data[features][:num_data])
y_new = np.array(new_data[outcome][:num_data])
```

```python
predictions_test = x_new.dot(parameters)

plt.scatter(x_new[:,1], y_new, alpha = 1, color='blue', label='actual')
plt.scatter(x_new[:,1], predictions_test, color='red', alpha = 0.5, label='predicted')
plt.legend(loc='upper right')
plt.title('Plot of PE against AT (test data)')
plt.ylabel('PE')
plt.xlabel('AT')
plt.show()
print 'The cost on the test data is:', cost_function(y_new[:num_data], predictions_test) 

plt.scatter(x[:num_data,1], y[:num_data], alpha = 1, color='blue', label='actual')
plt.scatter(x[:num_data,1], predictions_train, color='red', alpha = 0.5, label='predicted')
plt.title('Plot of PE against AT (training data)')
plt.legend(loc='upper right')
plt.ylabel('PE')
plt.xlabel('AT')
plt.show()
print 'The cost on the training data is:', cost_function(y[:num_data], predictions_train)
```

## Choosing the learning rate, $\alpha$

Choosing $\alpha$ can be tricky because:
1. If $\alpha$ is too large, the $\theta_j$ cannot be learnt.
* If $\alpha$ is too small, the algorithm wil run very slowly.

Here are 2 examples.

```python
print 'When the learning rate is too large.'
alpha = 6e-6 # 10 times less iteration then before. 6e-7 was the alpha used just now
parameters_, costs_ = linear_regression(x, y, initial_parameters, alpha, iter = 5000, show = None) 
plot_cost(range(len(costs_)), costs_, title = 'Plot of Cost against Iteration\nat alpha = %f' % alpha)
```

```python
alphas = [3e-7, 6e-7, 3e-8, 6e-8, 3e-9, 6e-9]
iterations = 5000
min_iter = 0

for alpha in alphas:
    initial_parameters = np.zeros(x.shape[1])
    parameters, costs = (linear_regression(x, y, initial_parameters, alpha = alpha
                                                            , iter=iterations, show = None))
    (plot_cost(range(min_iter, iterations), costs[min_iter:]
               , title='Plot of Cost Against Iteration\n(for different alphas)'
               , ylab='Cost', label = 'alpha = %s' %alpha))

plt.axis(ymax=1000)
plt.legend(loc='upper right')
plt.show()
```

If the learning rate is too small, the cost will take a long time to stabilize. However, the line between alpha being too small or too large is very fine.

## What else can we do to increase the speed of the algorithm?

### Perform data normalization.

$$mean = \bar{x} = \frac{1}{m}*\sum_{i=0}^m x_i$$
$$standard\ deviation = s = \sqrt{\frac{1}{m}\sum_{i=0}^m(x_i - \bar{x})^2}$$

For every $x$, $$x' = \frac{x - \bar{x}}{s}$$

```python
means = np.array(data[feats].apply(np.mean, axis = 0))
sd = np.array(data[feats].apply(np.std, axis = 0))

x_norm = data[['x_0']].copy()
x_norm[feats] = data[feats].apply(lambda x: (x - np.mean(x)) / np.std(x), axis = 1)
```

```python
print x_norm.head()
x_norm = np.array(x_norm)
```

```python
initial_parameters = np.ones(5)
alpha = 1e-6
alpha_norm = 1e-1
iterations = 5000

parameters_norm, costs_norm = linear_regression(x_norm, y, initial_parameters, alpha=alpha_norm, iter=iterations, show=None)

plt.plot(range(iterations), costs_norm, label = 'normalized')
plt.plot(range(iterations), costs[:iterations], label = 'un-normalized')
plt.axis(ymax = 400)
plt.legend(loc='upper right')
plt.show()
```

**Note:** It is not fair to compare performance between normalized and un-normalized in the above because the learning rate used is different.

But trust me that normalization is, in general, better ;)

### Pick better initial parameters.

We want to decrease the initial cost. Several strategies can be used.
1. Pick $\theta_0$ to be equal to the mean of the outcome.
    * So as to reduce the initial error of the model.
* Pick $\theta_j$ to be equal to the gradient of plotting each individual independent variables with the dependent variable.

#### Picking theta_0 to be equal to the mean of the outcomes.

```python
initial_parameters = np.ones(5)
initial_parameters[0] = np.mean(y)
print 'The initial parameters:', initial_parameters

parameters_meanOutcome, costs_meanOutcome = (linear_regression(x_norm, y, initial_parameters
                                                               , alpha=alpha_norm, iter=iterations, show=None))
    
plt.plot(range(iterations), costs_norm, label = 'normalized')
plt.plot(range(iterations), costs_meanOutcome, label = 'theta_0 = mean(y)')
plt.axis(ymax = 150, ymin=80)
plt.legend(loc='upper right')
plt.show()
```

The performance is similar to normalizing the data.

#### Picking theta to be equal to the gradient of plotting each individual independent variables with the dependent variable.

```python
def gradient_estimate(x, y):
    small = np.argmin(x)
    large = np.argmax(x)
    return (y[small] - y[large]) / (x[small] - x[large])
```

```python
initial_parameters = np.ones(5)
initial_parameters[0] = np.mean(y)

for i in range(x_norm.shape[1]):
    if i == 0:
        continue
    else:
        initial_parameters[i] = gradient_estimate(x_norm[:, i], y)
        
parameters_grad, costs_grad = (linear_regression(x_norm, y, initial_parameters
                                                 , alpha=alpha_norm, iter=iterations, show=None))
```

```python
(plot_cost(range(len(costs_grad))
            , costs_grad, title = 'Plot of Cost against Iteration\nat different alphas'
            , legend='upper right', label='init to grad'))
(plot_cost(range(len(costs_meanOutcome)), costs_meanOutcome, 
            title = 'Plot of Cost against Iteration\nat different alphas', legend='upper right', label='theta_0 = mean(y)'))
plt.axis(ymax=500)
```

There is no one size fit all method. You can try anything as long as the mathematics and the principles are not violated.
